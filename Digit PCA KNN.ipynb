```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
from sklearn.externals import joblib
```

```python
train_data = pd.read_csv('train.csv')
train_labels = train_data.loc[:,'label']
train_images = train_data.loc[:, 'pixel0':]
```

```python
pca = PCA(n_components=40, svd_solver='randomized',
          whiten=True).fit(train_images)

X_train_pca = pca.transform(train_images)
```

```python
knn = KNN(n_neighbors=3)
knn.fit(X_train_pca, train_labels)
```

```python
test_data = pd.read_csv('test.csv')
X_test_pca = pca.transform(test_data)
```

```python
results = knn.predict(X_test_pca)
```

```python
for item in results:
    print item
```
